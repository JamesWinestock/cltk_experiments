```python
from cltk.corpus.readers import get_corpus_reader
latin_corpus = get_corpus_reader(corpus_name = 'latin_text_latin_library', language = 'latin')
from cltk.stem.lemma import LemmaReplacer
lemmatizer = LemmaReplacer('latin')
```

```python
lower_livy_list = [[w.lower() for w in paras] for paras in livy_paras_flat]
```

```python
junk = ['.', ',', ' ', ':', ';', 'qui1', '-', 'que', '$', '%', '&','*','+', '-', '/', '<', '=', '>', '@', '^', '_',  '`', '{', '|', '}', '~', '?', '!', '«', '»']
junk_removed = [[w for w in paras if w not in junk] for paras in livy_paras_flat]
```

```python
lower_livy_list_no_junk = junk_removed
```

```python
def includes(*keyword):
    count = 0
    for item in flat_list:
        for para in item:
            #if keyword in para:
            #if any([keyword in para]) == True:
            if any(word in para for word in keyword):
                count += 1
                #break
    print(str(count) + " out of " + str(len(flat_list)) + " paragraphs")
```
